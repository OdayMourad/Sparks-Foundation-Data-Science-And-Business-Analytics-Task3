**1 ) Which features are available in the dataset:**


```python
print("Our dataset consist of {} samples".format(data.shape[0]))
print('-'*50)
print("There are {} features: {}".format(data.shape[1],data.columns.values))
print('-'*50)
```

```python
# We can dispense with the postal code feature
data = data.drop("Postal Code", axis=1)
```

```python
print("features possible values are :\n\n")
print('Ship Mode :\n')
print(data['Ship Mode'].value_counts())
print('-'*100)
print('Segment :\n')
print(data['Segment'].value_counts())
print('-'*100)
print('Country :\n')
print(data['Country'].value_counts())
print('-'*100)
print('City :\n')
print(data['City'].value_counts())
print('-'*100)
print('State :\n')
print(data['State'].value_counts())
print('-'*100)
print('Region :\n')
print(data['Region'].value_counts())
print('-'*100)
print('Category :\n')
print(data['Category'].value_counts())
print('-'*100)
print('Sub-Category :\n')
print(data['Sub-Category'].value_counts())
print('-'*100)
print('Sales :\n')
print(data['Sales'].value_counts())
print('-'*100)
print('Quantity :\n')
print(data['Quantity'].value_counts())
print('-'*100)
print('Discount :\n')
print(data['Discount'].value_counts())
print('-'*100)
print('Profit :\n')
print(data['Profit'].value_counts())
print('-'*100)
```
